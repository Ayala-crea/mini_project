```python
from pandas import *
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.stats import mode

import os
```

```python
penjualan = pd.read_csv('zara.csv')
```

```python
penjualan.head(5)
```

```python
# Pisahkan kolom berdasarkan delimiter ';'
data_split = penjualan.iloc[:, 0].str.split(';', expand=True)

# Beri nama kolom jika diperlukan
data_split.columns = [
    'Product ID', 'Product Position', 'Promotion', 'Product Category', 'Seasonal',
    'Sales Volume', 'brand', 'url', 'sku', 'name', 'description',
    'price', 'currency', 'scraped_at', 'terms', 'section'
]

# Tampilkan beberapa baris pertama dari data yang telah dipisahkan
data_split
```

```python
# Konversi kolom ke tipe data yang sesuai
data_split['Product ID'] = pd.to_numeric(data_split['Product ID'])
data_split['Sales Volume'] = pd.to_numeric(data_split['Sales Volume'])
data_split['price'] = pd.to_numeric(data_split['price'])
data_split['scraped_at'] = pd.to_datetime(data_split['scraped_at'])


# Tampilkan informasi tentang data setelah konversi
data_info = data_split.info()

data_info
```

```python
# Mendapatkan kolom numerik dalam rentang 2 hingga 15 (0-based index)
numerical_cols = data_split.iloc[:, 2:16].select_dtypes(include=[np.number]).columns

# Konversi kolom-kolom numerik tersebut ke 'float16'
data_split[numerical_cols] = data_split[numerical_cols].astype('float16')

# Tampilkan tipe data setelah konversi untuk verifikasi
print(data_split.dtypes)
```
