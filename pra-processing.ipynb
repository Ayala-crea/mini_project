```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('zara.csv')
df
```

```python
# Pisahkan kolom berdasarkan delimiter ';'
data_split = df.iloc[:, 0].str.split(';', expand=True)

# Beri nama kolom jika diperlukan
data_split.columns = [
    'Product ID', 'Product Position', 'Promotion', 'Product Category', 'Seasonal',
    'Sales Volume', 'brand', 'url', 'sku', 'name', 'description',
    'price', 'currency', 'scraped_at', 'terms', 'section'
]

# Tampilkan beberapa baris pertama dari data yang telah dipisahkan
data_split
```

```python
# Konversi kolom ke tipe data yang sesuai
data_split['Product ID'] = pd.to_numeric(data_split['Product ID'])
data_split['Sales Volume'] = pd.to_numeric(data_split['Sales Volume'])
data_split['price'] = pd.to_numeric(data_split['price'])
data_split['scraped_at'] = pd.to_datetime(data_split['scraped_at'])


# Tampilkan informasi tentang data setelah konversi
data_info = data_split.info()

data_info
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data_split[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data_split[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data_split[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'name', data = data_split, color = 'blue', 
              order = data_split['name'].head(10).value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'Product Category', data = data_split, color = 'blue', 
              order = data_split['Product Category'].head(10).value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'Product Position', data = data_split, color = 'blue', 
              order = data_split['Product Position'].head(5).value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'terms', data = data_split, color = 'blue', 
              order = data_split['terms'].head(15).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```
